```python
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
def funcao(param,x):
    return param[0] + param[1]*x
```

```python
df = pd.read_csv('exemplo2_line.dat', delim_whitespace=True)
df
```

```python
ele = pd.DataFrame(data = df.values, columns = ['x', 'y'])
ele
```

```python
x = ele['x']
y = ele['y']
```

```python
fun=funcao([4,0.5],x)
fun
```

```python
plt.plot(x, y,'.')
plt.plot(x,fun,label='Ajuste', color='red')
#plt.axvline(24000, color='r', linestyle='dotted', linewidth=2)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
fun_f=funcao([4,0.315],x)
plt.plot(x, y,'.')
plt.plot(x,fun_f,label='Ajuste', color='red')
#plt.axvline(24000, color='r', linestyle='dotted', linewidth=2)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
def min_quad(y,fun_f):
    S = np.zeros((len(fun_f)))
    for i in range(len(fun_f)):
        S[i] = (y[i] - fun_f[i])**2
    return S.sum()
```

```python
min_quad(y,fun_f)
```

```python
min_quad(y,fun)
```
