## Q1

###### Info:
 - Sigma = Desvio Padrão
 - Largura total à meia altura (FWHM) = 8 pixels
 - FWHM é medido por 2*sqrt(2*ln(2))*sigma 
 - (x0,y0) = (50,50)
 - Brilho = Bo * (e^((x-xo)^2 /2*sigma^2))/(sqrt(2*pi)*sigma) * (e^((y-yo)^2 /2*sigma^2))/(sqrt(2*pi)*sigma)
 - Cada coordenada segue uma distribuição normal
 - Fluxo total de 6.000 elétrons


```python
x0 = 0
y0 = 0
sigma = 8/(np.sqrt(8*np.log(2))) #Usando FWHM = 8pixel
soma=0

## Calculando a soma das exponenciais
for x in range(0,99):
    for y in range(0,99):
        soma = soma + np.exp((-x**2-y**2)/(2*(sigma**2)))
        
        
print(soma)
```

Sendo soma = 6000/B0, temos:
    B0 = 6000/22.64 = 265,017

A partir daqui, separamos os valores encontrados e criamos a matriz referente ao ceu inteiro. Na mesma matriz, logo em seguia, adicionamos o que deve ser a estrela, seguindo o calculo do brilho dada na questão:

```python
soma = 22.637
B0 = 265.017

# céu teórico com 100 eletrons por pixel
ceu = np.ones((100,100))*100

# Estrela teorica


def gaussiana(x, x0, y, y0): 
    return B0 * norm.pdf(x,x0,sigma)*norm.pdf(y,y0,sigma)



for j in range(100):
    for i in range(100):
        ceu[i][j] = gaussiana(i,50,j,50) 
#ceu[50][50] = estrela
```

Por fim, fazemos o plot:

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12,9))
axes[0].matshow(ceu)
axes[0].set_title('Ceu inteiro')
axes[1].matshow(ceu[50-10:50+10, 50-10:50+10])
axes[1].set_title('Recorte da estrela')
```

Criei uma função para aplicar a distribuição normal, de forma aleatória, em cada pixel do ceu. Aplicamos isso apenas numa outra variavel 'ceu_ruidu'. Isso porque o resultado não pareceu correto, e eu não quis comprometer a variável anterior.

```python
def normal(ceu): #definindo a função de distribuição normal
    sistema = []
    linhas = len(ceu) 
    colunas = len(ceu[0]) 
    for i in range(linhas):
        sistema.append([])
        for j in range(colunas):
            soma = ceu[i][j] + np.random.normal(0,5)
            sistema[i].append(soma)
    return sistema

ceu_ruidu = normal(ceu)
plt.matshow(ceu_ruidu)
```

### Q4

    - Formula geral: m = -2.5 * np.log10(Sestrela) + 10

```python
def magnitude(SestrelaAreal):
    return -2.5 * np.log10(abs(SestrelaAreal)) + 10

#Para a estrela real (com ruido)
m5 = magnitude(Sestrela5real)
m10 = magnitude(Sestrela10real)
m20 = magnitude(Sestrela20real)
m30 = magnitude(Sestrela30real)
m40 = magnitude(Sestrela40real)

print('\n','Para a estrela com ruído, a magnitude instrumental é:')
print('Na abertura de 5 =', m5)
print("Na abertura de 10 =", m10)
print("Na abertura de 20 =", m20)
print("Na abertura de 30 =", m30)
print("Na abertura de 40 =", m40)
```

```python
#Correções considerando m40 como magnitude instrumental de abertura

corr1 = abs(m5 - m40)
corr2 = abs(m10 - m40)
corr3 = abs(m20 - m40)
corr4 = abs(m30 - m40)

print('\n','Veremos qual abertura fornece a melhor estimativa da magnitude intrumental:\n')
print('Na abertura de 5 =', corr1)
print("Na abertura de 10 =", corr2)
print("Na abertura de 20 =", corr3)
print("Na abertura de 30 =", corr4)
```
