Primeiro vamos criar o céu, para isso precisamos criar uma matriz 100 pixels x 100 pixels onde cada pixel tem 100 elétrons, como a questão pede

```python
#Criando o céu
ceu = []
coluna = []
for j in range(100):
    coluna.append(100)
for i in range(100):
    i = i+1
    ceu.append(coluna)
ceu = np.array(ceu)
ceu #Criamos uma matriz 100 x 100 com valor de 100
```

```python
#Vamos confirmar que a matriz é de 100 x 100, portanto
print('Esta matriz tem ' + str(len(ceu[0])) + ' colunas e ' + str(len(ceu)) + ' linhas')
```

```python
#plotando a imagem da matriz céu
x = plt.figure(figsize = [16,9])
plt.imshow(ceu, interpolation = 'none',cmap='inferno')
plt.savefig('céu_sem_ruído.png')
plt.grid(True)
plt.title('Céu sem ruído \n',fontsize=15)
plt.colorbar(label='intensidade', orientation = 'vertical')
plt.show()
```

```python
#Encontrando a relação entre sigma e FWHM

f=8 #valor dado na questão
s = f/(2*((2*np.log(2))**(1/2)))

print(s)
```

```python
#encontrando o brilho gaussiano da estrela, para isso definiremos uma função
b0 = 6000 #valor informado na questão

#definindo a função brilho como gaussiano 
def brilho(b0,s,x,x0,y,y0):
    return b0* norm.pdf(x,x0,s)* norm.pdf(y,y0,s)
```

```python
#criando a estrela
estrela = np.zeros((100,100)) #criando uma matriz de 0 para fazermos um perfil de brilho gaussiano posteriormente

for j in range(100): #traçando o perfil gaussiano
    for i in range(100):
        estrela[i,j] = brilho(b0,s,i,50,j,50) #foi usado o 50 para centralizar a estrela 
```

```python
#plotando apenas a imagem da estrela
x = plt.figure(figsize = [16,9])
plt.imshow(estrela, interpolation = 'none',cmap='inferno')
plt.savefig('estrela.png')
plt.grid(True)
plt.title('Apenas a estrela \n',fontsize=15)
plt.colorbar(label='intensidade', orientation = 'vertical')
plt.show()
```

```python
#somando as matrizes céu e estrela para formamos o sistema ceu+estrela
def somar(ceu, estrela):
    ceu_estrela = []
    quant_linhas = len(ceu) 
    quant_colunas = len(ceu[0]) 
    for i in range(quant_linhas):
        ceu_estrela.append([])
        for j in range(quant_colunas):
            soma = ceu[i][j] + estrela[i][j]
            ceu_estrela[i].append(soma)
    return ceu_estrela

sistema = somar(ceu, estrela)
```

```python
x = plt.figure(figsize = [16,9])
plt.imshow(sistema, interpolation = 'none',cmap='inferno')
plt.savefig('céu com a estrela.png')
plt.grid(True)
plt.title('Céu com a estrela = Sistema\n',fontsize=15)
plt.colorbar(label='intensidade', orientation = 'vertical')
plt.show()
```
