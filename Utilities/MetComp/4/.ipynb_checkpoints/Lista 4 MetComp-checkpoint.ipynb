```python
star = np.array([np.loadtxt("estrelas/estrela_"+str(i)+".txt") for i in range(1,100+1)])
catalogo = pd.DataFrame(data = np.reshape(star,(195000,3)) ,columns = ['Comprimento de onda', 'Radiância espectral','Erro (radiância)'])
#Catalgo em Dataframe
star_compri, star_radi, star_errand = star.T  #Cada coluna do catalogo em array
catalogo.head()
```

```python
# variáveis
h=np.float(6.62607015*10**(-34))
c=np.float(299792458)
e=np.exp(1)
k=np.float(1.380649*10**(-23))
```

```python
def planck(lambd,T):
    return (((2*h*(c)**2))/(lambd**5))*(1/(e**((h*c)/(lambd*k*T))-1))

def chiquad(info, T):
    x, y, z = info.T                    #ideia: x = info['Comprimento de onda'], y=info['Radiância']... como d.f.                                                     
    return np.sum(np.power((y - planck(x,T))/z,2))

def best(estrela, mini=1000, maxi=50000):
    chimenor = float('inf')
    tempe = 0
    tmin, tmax = mini, maxi
    while True:
        temps = [round(i,0) for i in np.linspace(tmin, tmax, 20)]
        chi = chimenor
        for temp in temps:
            z = chiquad(star[estrela-1], temp) #catalogo['Comprimento de onda'][estrelas-1]
            if z < chimenor:
                chimenor, tempe = z, temp
        if chimenor == chi or chimenor < 1.0e-5:
            break
        tmin = temps[list(temps).index(tempe) - 1]
        if tmin == temps[-1]:
            tmin = temps[0]
        tmax = temps[list(temps).index(tempe) + 1]
    return tempe
#Ideia de utilizar DataFrame Catalogo invés da array acabou se mostrando muito mais complicada de se aplciar na função
def classif(T):
    if T >= 33000:
        return 'O'
    if T >= 10000:
        return 'B'
    if T >= 7500:
        return 'A'
    if T >= 6000:
        return 'F'
    if T >= 5200:
        return 'G'
    if T >= 3700:
        return 'K'
    else:
        return 'M'
    
lista = []

tempos = time.time()
for f in range(1,100+1):
    te = time.time()
    m = best(f)
    l = classif(m)
    
    lista.append({'Estrela': f, 'Temperatura (K)': m, 'Classificação': l, 'Tempo': time.time() - te})

print ('Tempo total: '+ str(time.time()-tempos))
for c in lista:
    print(c)
```

```python
def wien(pontos):
    c, r, e = pontos.T
    pico = c[r.argmax()]
    return 0.0028976 / pico

estrelas = []
for e in range(1, 101):
    estrelas.append({'Estrela': e, 'Temperatura(K)': wien(star[e-1])})
    print(estrelas[e-1])
```

2- [Paralelização e algorítimos iterativos 2D]

```python
#No arquivo as colunas são ângulo (em graus), raio (em UA) e erro do raio (em UA), onde o raio é medido com relação ao Sol*
orbita = np.loadtxt("./orbita_de_um_asteroide.dat")
ang = orbita[:,0] #Angulo graus
raio = orbita[:,1] # raio em UA
err = orbita[:,2] # erro raio
plt.rcParams['axes.facecolor'] = 'snow'
plt.polar(ang*np.pi/180,raio,'yo') # passando angulo em graus pra rad. (yo -> amarelo em pontinhos)
plt.show()
```

```python
a1 = 50
a2 = 20
a3 = 0
b1 = 40
b2 = 10
b3 = 0

def e(a,b): #função da excentricidade
    return np.sqrt(a**2-b**2)/a 

while (2*abs((a1-a2)/(a1+a2))> 0.0005) and (2*abs((b1-b2)/(b1+b2))> 0.0005):
    valores_aprox = []
    valores_a = np.arange(a2,a1,(a1-a2)/10)
    valores_b = np.arange(b2,b1,(b1-b2)/10)
    for i in valores_a:
        for j in valores_b:          #para cada valor de a teste inumeros valores de b
            ex = e(i,j)        # excentricidade da elipse
            aprox = sum(((raio - i*(1-ex**2)/(1+ex*np.cos(ang*np.pi/180)))/err)**2) #valores aproximados utilizando a equação do raio com origem em um dos focos
            valores_aprox.append(aprox)
    for i in range(10):
        for j in range(10):
            if valores_aprox[10*i+j] == min(valores_aprox):
                a3 = valores_a[i]
                a2 = valores_a[i-1]
                a1 = valores_a[i+1]
                b3 = valores_b[j]
                b2 = valores_b[j-1]
                b1 = valores_b[j+1]
print('a=',a3)
print('b=',b3)



angle = np.arange(0,2*np.pi,np.pi/180)
plt.polar(angle,a3*(1-e(a3,b3)**2)/(1+e(a3,b3)*np.cos(angle)))
plt.polar(ang*np.pi/180,raio,'r^')
plt.show()


a3min = 0.85*a3
a3max = 1.15*a3
b3min = 0.85*b3
b3max = 1.15*b3

print ('\n Plot final: \n\n')
plt.rcParams['axes.facecolor'] = 'black'
plt.figure(figsize=(7,7))
plt.polar(angle,a3*(1-e(a3,b3)**2)/(1+e(a3,b3)*np.cos(angle)),color='aqua')
plt.polar(0,0,'yo')
plt.polar(ang*np.pi/180,raio,'r.')
plt.polar(angle,a3min*(1-e(a3min,b3min)**2)/(1+e(a3min,b3min)*np.cos(angle)),color='lightseagreen',alpha=0.5)
plt.polar(angle,a3max*(1-e(a3max,b3max)**2)/(1+e(a3max,b3max)*np.cos(angle)),color='lightseagreen',alpha=0.5)
plt.fill_between(angle,a3min*(1-e(a3min,b3min)**2)/(1+e(a3min,b3min)*np.cos(angle)), a3max*(1-e(a3max,b3max)**2)/(1+e(a3max,b3max)*np.cos(angle)),color='cyan',alpha=0.2)
plt.show()
```
