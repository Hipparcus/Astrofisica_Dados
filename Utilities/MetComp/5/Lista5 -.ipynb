```python
def mc_integral_simple(func, xmin, xmax, npoints=100000):
    base = (xmax - xmin)
    x_sample = np.random.uniform(low=xmin, high=xmax, size=(npoints,))
    y_sample = np.apply_along_axis(func,0,x_sample)
    return np.mean(y_sample) * base

def f3a(x):
    return -1+np.sqrt(x)

def f3b(x):
    return np.exp(-x)
```

```python
x = np.arange(0, 3+1/1000, 1/1000)
ya = np.apply_along_axis(f3a, 0, x)
yb = np.apply_along_axis(f3b, 0, x)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12,4))
axes[0].plot(x, ya, c='m')
axes[0].fill_between(x, ya, 0, color='m', alpha=0.3)
axes[0].grid()
axes[1].plot(x, yb, c='c')
axes[1].fill_between(x, yb, 0, color='c', alpha=0.3)
axes[1].grid()
plt.tight_layout()
plt.show()

ta = time()
inta = mc_integral_simple(f3a, 0, 3)
tb = time()
intb = mc_integral_simple(f3b, 0, 3)
end = time()

print(tb-ta, end-tb)

print(f'Integral da função a: {inta}')
print(f'Integral da função b: {intb}')
```

```python
def integrate(x1,x2,func,n=100000):
    """Retorna, nesta ordem:
    -> Resultado da integral
    -> x dos pontos
    -> y dos pontos
    -> Uma lista que diz se os pontos estão entre a função e 0, indicando se acima ou abaixo do 0, ou não"""
    X=np.linspace(x1,x2,1000) #lista de 1000 pontos entre x1 e x2
    y1=min(func(X))-0.5 #y menor que o menor valor encontrado para a função
    y2=max(func(X))+0.5 #y maior que o maior valor encontrado para a função
    area=(x2-x1)*(y2-y1) #area do retângulo contendo pontos
    check=[] #lista de pontos válidos (-1 ou 1 se estiver entre a função e 0)
    xs=[]
    ys=[]
    for i in range(n):
        x=np.random.uniform(x1,x2,1) #x aleatorio
        xs.append(x)
        y=np.random.uniform(y1,y2,1) #y aleatorio
        ys.append(y)
        if (fx:=func(x)) > 0: #se o y da função estiver acima de 0
            if y>fx or y<0: #se o y do ponto não estiver entre a funçao e 0
                check.append(0) #adiciona 0 na lista de pontos válidos
            else: #se o ponto estiver entre a função e 0
                check.append(1) #adiciona 1 na lista de pontos válidos
        else: #se o y da função estiver abaixo de 0
            if y<fx or y>0: #se o y do ponto não estiver entre a função e 0
                check.append(0) #adiciona 0 na lista de pontos válidos
            else: #se o ponto estiver entre a função e 0
                check.append(-1) #adiciona -1 na lista de pontos válidos
    return(np.mean(check)*area,xs,ys,check)
```

```python
ta = time()
ra,xa,ya,ca = integrate(0, 3, f3a)
ea = time()
df3a = pd.DataFrame()
df3a['x'] = xa
df3a['y'] = ya
df3a['c'] = ca

print(ea-ta)

tb = time()
rb,xb,yb,cb = integrate(0, 3, f3b)
eb = time()
df3b = pd.DataFrame()
df3b['x'] = xb
df3b['y'] = yb
df3b['c'] = cb

print(eb-tb)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12,4))
axes[0].plot(x,f3a(x))
axes[0].scatter(df3a[df3a['c']==0]['x'], df3a[df3a['c']==0]['y'], color='red')
axes[0].scatter(df3a[abs(df3a['c'])==1]['x'], df3a[abs(df3a['c'])==1]['y'], color='blue')
axes[0].grid()

axes[1].plot(x,f3b(x))
axes[1].scatter(df3b[df3b['c']==0]['x'], df3b[df3b['c']==0]['y'], color='red')
axes[1].scatter(df3b[abs(df3b['c'])==1]['x'], df3b[abs(df3b['c'])==1]['y'], color='blue')
axes[1].grid()

plt.tight_layout()
plt.show()
```
