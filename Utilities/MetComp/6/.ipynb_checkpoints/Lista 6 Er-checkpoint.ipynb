```python
import numpy as np
import matplotlib.pyplot as plt

#import scipy.integrate as integrate
import scipy as sci
import scipy.special as special

from scipy.misc import derivative



import sympy as sym
from sympy import symbols, Eq, solve
from sympy import *
```

```python
#1 1. [Derivação] Encontre a derivada da função arctan(x^2) utilizando o sympy, scipy (para o scipy obtenha os valores 
#da derivada com x variando de 0 até 4) e numpy (para o numpy gere um array com x variando de 0 até 4 em 10^6 passos)

#Simpy

x = sym.Symbol('x')

der0 = sym.diff(sym.atan(x**2))

f0 = Derivative(sym.atan(x**2))

eq = Eq(f0, der0) 

print ('A derivada da função é:')

eq
```

```python
#Scipy -> com x variando de 0 até 4

def f0(x):
    return np.arctan(x**2)

scites = [derivative(f0, i, dx=1e-6) for i in range(5)]
sci0 = [round(num, 2) for num in scites]


print("A derivadas pelo Scipy de 0 até 4 são:", sci0)
```

```python
#Numpy

#Vou usar a função f0 ali de cima

arr = np.linspace(int(0),int(4),int(10**6))

dx = np.diff(f0(arr))

arr1 = arr[0:999999]

print ('Plot da derivada da função:')

plt.plot(arr1, dx, label='Derivada')
plt.legend()
plt.show()
```

```python
#3 Resolva usando o Sympy e o Scipy a seguinte equação diferencial:
# y''+2y'+3y=sin(x^2), sendo y(0)=0 e y'(0)=0

#Sympy

#y = sym.Symbols('y', cls= sym.Function)
t = sym.Symbol('t')
y = sym.symbols('y', cls = sym.Function)

#y1 = sym.dsolve(sym.diff(y(x),x,x) + 2*sym.diff(y(x), x) + 3*y(x),sym.sin(x))

edo = sym.Eq(sym.diff(y(t),t,t) + 2 * sym.diff(y(t),t) + 3 * y(t), sym.sin(t))

print("A EDO é:")
edo
```

```python
#Resolvendo o EDO
print('Resolvendo a EDO:')
sEDO = sym.dsolve(edo)
sEDO
```

```python
#Agora é preciso calcular com as constantes

C = sym.solve([sEDO.rhs.subs(t,0) - 0,sEDO.lhs.diff(t).subs(t,0) - 0])

#Põe as constantes de volta no EDO
sEDO2 = sEDO.subs(C)

#Transforma em algo calculável
sEDO3 = sym.lambdify(t,sEDO2.rhs, 'numpy')

#Fazer um array pra plotar
t = np.arange(0,20,0.1)

plt.plot(t,sEDO3(t))

#Aqui supostamente deveria plotar o mesmo gráfico da questão abaixo, mas por algum motivo não quer funcionar
#Então deixo aqui minha tentativa
```

```python
#Scipy

#Pra facilitar o entendimento 
#z = y' -> z' + 2z + 3y = sin(x^2) | y(0)= z(0)= 0

from scipy.integrate import odeint

def fun(V,x):      #V é um vetor tal que y=V[0] e z=V[1]
                    
    return [V[1], - 2*V[1] - 3*V[0] + np.sin(x**2)]

V0 = [0,0]

xs = np.linspace(0,20,200)

Vs = odeint(fun, V0, xs)

ys = Vs[:,0]
```

```python
#Plotando

plt.plot(xs,ys)
plt.title("EDO - 2ª Ordem")
plt.show()
```

```python
#4)a)Gere uma matriz aleatória quadrada de 3x3 de números inteiros positivos. Encontre os autovetores e autovalores e com isso
#a matriz diagonalizada b) Calcule o determinante c) Prove que o determinante é o volume gerado do paralelepípedo pelos
#autovetores (como arestas) e pelas linhas originais da matriz não diagonalizada, comparando com o produto escalar 
#triplo destes vetores.

#a)Gerar uma matriz 3x3 aleatória encontrar os autovalores e autovetores, e com isso a matriz diagonalizada

#Tem que usar randint
from sys import maxsize

Am = np.random.randint(0,maxsize,size=(3,3),dtype='int64')

aval, avet = np.linalg.eig(Am)
print ('A matriz é:')
print (Am)

print ('\n')

print ('Os autovalores são:', aval)
print ('Os autovetores são:')
print (avet)
print('\n')

#Para diagonalizar a matriz
diag = np.diag(aval)
print('A matriz diagonalizada:')
print (diag)
```

```python
#b)Calcular o determinante da matriz

detA = abs(round(np.linalg.det(Am),2))
print ('O determinante de A é:', detA)
```

```python
#c)Provar que o determinante é o volume gerado do paralelepípedo pelos autovetores...

#Definir cada um dos autovetores
av1, av2, av3 = avet[0],avet[1],avet[2]

#Somar a matriz com os autovetores e descobrir as arestas do sólido

for v in [av1, av2, av3]:
    arest = (Am) + (v)
    
a1, a2, a3 = arest

#Produto escalar triplo/descobrir o volume

volume = abs(np.round(np.dot(np.cross(a1, a2), a3)))

print ('O volume é:', volume)
print ('O determinante é:', detA)

#Curiosidade: não consegui fazer o volume e o determinante ficarem iguais sem ser usando o 'maxsize' lá em cima
```

```python
#5) Leia os arquivos amostra1.dat e amostra2.dat, encontre a curva interpolada com Spline no caso 1 e Spline Cúbica no caso 2,
#utilize as funções de ajuste de Spline do scipy. Faça o plot comparando a curva e os dados de entrada.

from scipy import interpolate

#Carregar o amostra1

amo1 = np.loadtxt('amostra1.dat')
err1 = np.loadtxt("erros1.dat")

plt.plot(amo1)
plt.title('Amostra 1')
plt.show()
```

```python
#Agora interpolando

x1 = np.linspace(0,10000, num = amo1.size, endpoint = True)

y1 = interpolate.splrep(x1,amo1,k=1,s=20)

inter1 = interpolate.splev(x1, y1)

#plt.plot(x1,y1)
plt.figure(figsize=(10,8))
plt.plot(x1, amo1,':', x1, inter1,'-')
plt.title('Interpolação da Amostra1')
plt.legend(['Dados Amostra1', 'Interpolação/spline'])
plt.show()
```

```python
#Fazer a amostra2 agora

amo2 = np.loadtxt('amostra2.dat')
err2 = np.loadtxt("erros2.dat")

plt.plot(amo2)
plt.title('Amostra 1')
plt.show()
```

```python
#Agora interpolar a amostra 2

#Pegando os valores da amostra 2 e fazer um x de mesmo tamanho para "curva linear"

y2 = amo2[4000:6000:250]
x2 = np.arange(4000,6000,250)

#Interpolação Cúbica
inter2 = interpolate.splrep(x2, y2, k=3)

# X para interpolar
x3 = np.arange(4000, 6000, 1)

# Y para interpolar
y3 = interpolate.splev(x3, inter2, der=0)

plt.figure(figsize=(10,8))
plt.plot(x2, y2,'--', x2, y2, 'o', x3, y3)
plt.legend(['Linear', 'Pontos', 'Interpolação Cúbica'])
plt.title('Amostra 2 - Interpolação Cúbica')
plt.show()
```

```python
#Com as informações aqui eu volto pra fazer a questão 2 

#Temos que integrar as duas amostras
#Amostra1

int_amo1 = sci.integrate.quad(lambda x: interpolate.splev(x,y1,), -2, 4)[0]

print('A integral da Amostra 1 de -2 até 4 é:', round(int_amo1,2))
```

```python
#Amostra 2

int_amo2 = sci.integrate.quad(lambda x: interpolate.splev(x,inter2,), -2, 4)[0]

print("A integral da amostra 2 de -2 até 4 é:", round(int_amo2,2))
```
