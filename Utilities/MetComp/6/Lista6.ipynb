    3-[Equações diferenciais]

```python
#Sympy

t = sym.Symbol('t')
y = sym.symbols('y', cls = sym.Function)


edo = sym.Eq(sym.diff(y(t),t,t) + 2 * sym.diff(y(t),t) + 3 * y(t), sym.sin(t))

print("A EDO:\n")
edo
```

```python
print('Solução:\n')
sol = sym.dsolve(edo)
sol
```

```python
#Scipy

def fun(V,x):  #V como vetor em que y=V[0] e z=V[1]
                    
    return [V[1], - 2*V[1] - 3*V[0] + np.sin(x**2)]

V_zero = [0,0]

x = np.linspace(0,20,200)

Vs = odeint(fun, V_zero, x)

y = Vs[:,0]

#Plotando

plt.plot(x,y,color='y')
plt.title("EDO - 2ª Ordem")
plt.show()
```

```python
from sys import maxsize

Am = np.random.randint(0,maxsize,size=(3,3),dtype='int64')

aval, avet = np.linalg.eig(Am)
print ('A matriz é:')
print (Am)

print ('\n')

print ('Autovalores :\n'+ str(aval))
print('\n')
print ('Autovetores :\n'+str(avet))
print('\n')


diag = np.diag(aval) #diagonalizar matriz
print('Matriz diagonalizada:\n ' + str(diag))
```

```python
am1 = np.loadtxt("./amostra1.dat")
am2 = np.loadtxt("./amostra2.dat")
err1 = np.loadtxt("./erros1.dat")
err2 = np.loadtxt("./erros2.dat")
```

```python
x1 = np.linspace(-2,4,num=am1.size)
tck1 = interpolate.splrep(x1,am1,k=1,s=20)
y1 = interpolate.splev(x1, tck1)

plt.figure(figsize=(12,8))
plt.subplot(2,1,1)
plt.plot(x1,am1) #plot amostra1
plt.subplot(2,1,2)
plt.plot(x1,y1,'red')
plt.show()
```

```python
x1 = np.linspace(0,10000, num = am1.size, endpoint = True)

y1 = interpolate.splrep(x1,am1,k=1,s=20)

inter1 = interpolate.splev(x1, y1)

#Interpolando
plt.plot(x1, am1,':', x1, inter1,color='r')
plt.title('Interpolação da Amostra1')
plt.legend(['Dados Amostra1', 'Interpolação/spline'])
plt.show()
```
